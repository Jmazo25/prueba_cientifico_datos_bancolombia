# Carga y Parseo del csv

```python
nombres = ['INFORMACIÓN BASICA_Primer Nombre', 'INFORMACIÓN BASICA_Segundo Nombre','INFORMACIÓN BASICA_Primer Apellido',
        'INFORMACIÓN BASICA_Segundo Apellido', 'INFORMACIÓN BASICA_Sexo']
df_result = pd.DataFrame(columns= nombres)
df_result
```

```python
# nombres = {'INFORMACIÓN BASICA_Primer Nombre', 'INFORMACIÓN BASICA_Segundo Nombre','INFORMACIÓN BASICA_Primer Apellido',
#         'INFORMACIÓN BASICA_Segundo Apellido', 'INFORMACIÓN BASICA_Sexo'}
# df_result = pd.DataFrame(columns= nombres)

with open('RUAF.csv',newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=';')
    total1 = []
    conc_dict ={}
    for row in reader:
        if row[0] == '¬-*-¬DATA':
            total1.append(conc_dict)
            conc_dict = {}
        elif row[0] == '¬-*-¬ID':
            conc_dict['key_fep'] = row[1]
        elif row[0] != '¬-*-¬ID' and row[-2] != '' and row[2] != 'La consulta no fue exitosa':
            key_n = row[2] + '_'+row[3]
            conc_dict[key_n] = row[-2]
            conc_dict['num_id'] = row[1]
            conc_dict['tipo_id_str'] = row[0][-2:]
            conc_dict['f_consulta'] = row[-1]
        elif row[2] == 'La consulta no fue exitosa':
            reg_died = {}
            reg_died['num_id'] = row[1]
            reg_died['f_consulta'] = row[-1]
            reg_died['tipo_id_str'] = row[0][-2:]
            reg_died['marca_sin_informacion'] =1
            total1.append(reg_died)
    total1.remove({})
    for dict_r in total1:
        #print(dict_r)
        df_temp = pd.DataFrame(dict_r,index=[0])
        df_result = pd.concat([df_result,df_temp],ignore_index=True)
```

Se crea un DataFrame de pandas sin valores y con pocas columnas, sólo para tenerlo listo cuando vaya ser anexada la data, luego se carga el archivo desde la misma carpeta donde se encuentra el notebook y con csv.reader se pasa a leer las líneas contenidas mediante un ciclo 'for', consecutivamente se crean un diccionario para cada registro para luego anexarlo como una línea al dataframe con la función append.

los diccionarios se crean a partir de ciertas condiciones validando las posiciones de las líneas y así efectuando ciclos de decisión como  el inicio de registro, la clave donde está el ID del mismo y si el registro no contienen estos valores anteriores inicia armado clave valores con secciones_campos para claves y el valor para valores, los demas campos como num_id_, f_consulta, tipo_id_str se crean a partir de los datos de las dos primeras posiciones que ya tienen este tipo de líneas. para los que no tienen infoprmación se crean una condición aparte para que se agregue la marca de no tienen información.

todos los diccionario luego son alamacenados en una lista y posteriormente recorridos para ser anexados como se comenta en el primer parrafo.

# Merge de los campos con fechas de afiliaciones a salud

```python
flag_ks = ['True' for k_s in list(df_result.columns) if k_s == 'f_afiliacion_salud_']
```

```python
flag_ks_2 = ['True' for k_s2 in list(df_result.columns) if k_s2 == 'f_afiliacion_salud_d']
```

```python
if flag_ks and flag_ks_2:
    df_result['f_afiliacion_salud'] = df_result.f_afiliacion_salud_ + df_result.f_afiliacion_salud_d
    df_result = df_result.drop(columns=['f_afiliacion_salud_','f_afiliacion_salud_d'])
elif flag_ks:
    df_result['f_afiliacion_salud'] = df_result['f_afiliacion_salud_']
    df_result = df_result.drop(columns=['f_afiliacion_salud_'])
else:
    df_result['f_afiliacion_salud'] = df_result['f_afiliacion_salud_d']
    df_result = df_result.drop(columns=['f_afiliacion_salud_d']) 
```
